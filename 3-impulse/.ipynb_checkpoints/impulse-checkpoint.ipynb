## Detectar spike en isis

```python
#Input: Pulse events 1, pulse events 2, spikes
#Output: array of index of events where there is a spike.
def check_spike(a,b,spikes):
    sp_indexes = []
    for i in range(a.shape[0]-1):
        #Guarda indices donde se de spike.
        elems = spikes[np.where((spikes > a[i]) & (spikes < b[i]))]
#         print(a[i],b[i],elems)
        if len(elems) >0 :
#             plt.plot(isi1[i],isi2[i],'.',color='blue')
            sp_indexes.append(i)
        
    return sp_indexes
```

```python
def plot_isi_cond(isi1,isi2,indexes,title,color='blue',label='',acc_decc=True):
    lim = max(max(isi1),max(isi2))+0.5
    #accelerating
    
    isi1 = isi1[indexes]
    isi2 = isi2[indexes]
    
#     isi1_1 = isi1[np.where(isi1 >= max(isi1)/2)]
#     isi1_1 = isi1[np.where(isi2 <= max(isi2)/2)]
#     isi2_1 = isi2[np.where(isi1 >= max(isi1)/2)]
#     isi2_1 = isi2[np.where(isi2 <= max(isi2)/2)]
    
#     #deccelerating
#     isi1_2 = isi1[np.where(isi1 < max(isi1)/2)]
#     isi2_2 = isi2[np.where(isi2 > max(isi2)/2)]
    
# #     plt.plot(isi1_1,isi2_1,'.',label=label)
#     plt.plot(isi1_2,isi2_2,'.',label=label)

    if(acc_decc):
#         col = np.where(isi1>=isi2,'g','r')
#         label = 'Accel. Green\n Decel. Red'
#         label = col
        x_1 = isi1[np.where(isi1>=isi2)]
        y_1 = isi2[np.where(isi1>=isi2)]
        
        x_2 = isi1[np.where(isi1<isi2)]
        y_2 = isi2[np.where(isi1<isi2)]
        
        scatter =  plt.scatter(x_1,y_1,c='firebrick',label="Dec.")
        scatter =  plt.scatter(x_2,y_2,c='forestgreen',label="Acc.")
        plt.legend()
    else:
#         col = np.full(isi1.shape,color)
        scatter =  plt.scatter(isi1,isi2,c=color,label=label)
        

        
    plt.xlabel('ISI1 (s)')
    plt.ylabel('ISI2 (s)')
    plt.xlim(0,lim)
    plt.ylim(0,lim)
    plt.title(title)
    if(label!=''):
        plt.legend()
```

## Spikes segÃºn pulsos

```python
isi1 = second_p-first_p
isi2 = third_p-second_p
ibi = first_p[1:] -third_p[:-1]
# print(third_p[0],first_p[1])

indexes_1 = check_spike(first_p,second_p,spikes)
indexes_2 = check_spike(second_p,third_p,spikes)
indexes_3 = check_spike(third_p[:-1],first_p[1:],spikes)
indexes_3= np.array(indexes_3)
```

```python
plt.figure(figsize=FIGSIZE)
plt.subplot(1,3,1)

plot_isi_cond(isi1,isi2,indexes_1,"Spike after 1st")

plt.subplot(1,3,2)

plot_isi_cond(isi1,isi2,indexes_2,"Spike after 2nd",'green')

plt.subplot(1,3,3)

plot_isi_cond(isi1,isi2,indexes_3,"Spike after 3rd",'red')


plt.savefig(path+"spikes_per_pulse.png")

# plt.show()
# plt.plot(isi1,isi2,'.')
```

## Stats

```python
def get_acc_dec(elems,i1,i2):
    elems = np.array(elems)
    acc = indexes_2[np.where(i1[elems]>=i2[elems])]
    dec = indexes_2[np.where(i1[elems]<i2[elems])]
    return acc,dec
```

```python
def plot_stats(elems,title,labels):
    plt.title(title)
    x = range(len(elems))
    y = [len(elem) for elem in elems]
    plt.bar(x,y,width=0.1,label=labels)
```

```python

plot_stats([indexes_1,indexes_2,indexes_3],"Spikes after each pulse",['1st','2nd','3rd'])
```

```python
ad2 = get_acc_dec(indexes_2,isi1,isi2)
ad3 = get_acc_dec(indexes_3,isi1,isi2)

plot_stats(ad2,"Spikes after 2nd",["Acc.","Dec."])
plot_stats(ad3,"Spikes after 3nd",["Acc.","Dec."])
```
